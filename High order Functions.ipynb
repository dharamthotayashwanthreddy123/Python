```python
n=int(input("Enter number of values in list: "))
C=[]
for i in range(n):
    value=int(input("Enter the value: "))
    C.append(value)
F=list(map(lambda x: (x*9/5)+32,C))
print(F)
```

```python
from functools import reduce

l=[1,2,3,4,5]
print(reduce(lambda x,y: x*y,l))
```
